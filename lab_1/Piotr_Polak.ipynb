```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
```

```python
def foo(x):
    return x*x +5
x1=np.linspace(-1,1,1000)
x2=np.linspace(-6,6,1000)
x3=np.linspace(0,5,1000)
```

```python
plt.figure(figsize=(10,10))

plt.plot(x2,foo(x2),label="przedzial=<-6,6>")
plt.plot(x3,foo(x3),label="przedzial=<0,5>")
plt.plot(x1,foo(x1),label="przedzial=<-1,1>",color="black",linestyle="-")
plt.legend(loc=4)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("f=x*x + 5")
plt.grid()
plt.show()
```
